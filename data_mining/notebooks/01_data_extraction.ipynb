```python
from sqlalchemy import create_engine, inspect
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
engine = create_engine(
    f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}/{os.getenv('DB_NAME')}"
)
```

Extracting Data From All Tables 


```python
# getting all the tables in star schema
inspector = inspect(engine)
tables = inspector.get_table_names(schema='warehouse')
# exporting all the tables to csv files
for table in tables:
  df=pd.read_sql(f"SELECT * FROM warehouse.{table}", engine)
  raw_folder = "../data/raw"
  csv_path = os.path.join(raw_folder, f"{table}.csv")
  df.to_csv(csv_path, index=False)
  print(f"Exported {table} → {csv_path}")


print("All tables extracted to CSV!")
```
